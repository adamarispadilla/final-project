1 Auto Parts

```python
import pandas as pd                 ##used for data analysis 
import numpy as np                  ##used for mathematical operations on arrays 
##association rule algorithm
from mlxtend.frequent_patterns import apriori, association_rules 
```

```python
##Reads txt file that contains data that will be analyzed and converts it into 
##a csv file. Then it gets transformed into a dataframe. 

read_file=pd.read_csv(r'autoparts.txt', sep='\t')
read_file.to_csv(r'autoparts.csv', index=None)
df=pd.read_csv('autoparts.csv')
df.head()        ##shows the first five rows of the dataframe 
```

```python
##counts missing data 
print(np.count_nonzero(df.isnull()))
```

```python
#add a column header 'Auto Parts' 
df=pd.read_csv('autoparts.csv', names=['Auto Parts'])
df.head() 
```

```python
##states the rows and columns of the dataframe
##This dataframe has 25 rows and 1 column
df.shape
```

```python
##Splits the products into a new format 

data=list(df["Auto Parts"].apply(lambda x:x.split(",")))
data
```

```python
#Transform data into a logical dataframe. Each column represents an 
##item and each row represent a record or a transaction for one 
##purchase. TRUE if the autopart transaction exists FALSE if not.

from mlxtend.preprocessing import TransactionEncoder
a=TransactionEncoder()
a_data=a.fit(data).transform(data)
df=pd.DataFrame(a_data, columns=a.columns_)

#Turn False & True to 0 & 1 for easy readability
df=df.replace(False,0)
df=df.replace(True,1)

df
```

```python
#set a threshold value for the support value (0.2) and calculate the support value.

#The support value can be set at any value you desire.
#use_colnames=True is to convert the integer values back into its respective item names.
df=apriori(df,min_support=0.2, use_colnames=True)
df
```

From above result, we say there are total we got 287 rules with 0.20 minimum support. Minimum support means the number of transactions in which x appears divided by the total number of transactions.

For example, oil has a support of 0.64 which means that out of all the autoparts, oil appears or gets bought 64% of the time. 

```python
#Let's view our interpretation values using the Association rule function for lift. 
df_ar=association_rules(df,metric="lift", min_threshold=1)
df_ar
```

From above result, we say that the support is 0.36 calculated by dividing the number of transactions containing oil and breakpad by the total number of transactions. The confidence level is 0.56 shows that out of all the transactions that contain breakpads 0.56 contain oil too. The lift 1.08 tells us that breakpad is 1.08 times more likely to be bought by the customer who also buys oil.

```python
#Let's view our interpretation values using the Association rule function for confidence.
df_ar=association_rules(df,metric="confidence", min_threshold=0.6)
df_ar
```

From above result, let's take a look at the first row items: breakpad and engine. The two items have the support of 0.32, where breakpad is 0.52 and engine is 0.68. confidence is 0.62, this implies that if a person bought brakepads then there is 62% engine is also bought. There are also higly positive correlation as we can see from the positive values of lift and conviction.
