```python
##importing required libraries 
import pandas as pd
import numpy as np

df=pd.read_csv("Brain Tumor.csv")
df

# Class column defines either the image has tumor or not (0 = Tumor, 1 = Non-Tumor)
```

```python
df.info()
```

```python
df.shape
```

```python
##dropping any null values just to double-check
df = df.dropna()
df.info()
```

```python
#rows and columns after dropping the columns with null values 
df.shape
```

```python
#In order to process we must turn those categorical values into numberic values
#Import LabelEncoder
from sklearn import preprocessing

#creating labelEncoder
le = preprocessing.LabelEncoder()

#Using .fit_transform function to fit label
#encoder and return encoded label
Skewness_label=le.fit_transform(df['Skewness'])

#printing skewness_label
Skewness_label
```

```python
#repeat it for Variance label
Variance_label=le.fit_transform(df['Variance'])

#printing Variance_label
Variance_label
```

```python
#repeat it for class label
class_label=le.fit_transform(df['Class'])

#printing class_label
class_label
```

```python
#we'll use zip method in this function called merge to combine the two labels: Skewness and Variance
def merge (list1, list2):
  merged_list=tuple(zip(list1, list2))
  return merged_list

features=merge(Skewness_label, Variance_label)
features
```

```python
#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
model = GaussianNB()

# Train the model using the training sets
model.fit(features, class_label)

#Predict 
predicted= model.predict([[2565, 1837]]) ##2565 = 4.276477026 skewness , 1837 = 619.5878446 varience 
print ("Predicted Value:", predicted)

##The model predicts that the person has a tumor. 
```

After encoding the Skewness and Varience levels of the image, we get a set of numbers. We use those numbers in the Guassian Classifier. We plugged in 2565 (4.3 skewness), 1837 (619.6 varience) which resulted in 1. This indicates whether the person does not have a tumor based on the image scan. The following values indicate whether a person has a tumor or not.

1 = No Tumor 

0 = Tumor

```python
df.colums = [
 'Class',
 'Mean',
 'Variance',
 'Standard Deviation',
 'Entropy',
 'Skewness',
 'Kurtosis',
 'Contrast',
 'Energy',
 'ASM',
 'Homogeneity',
 'Dissimilarity',
 'Correlation',
 'Coarseness']

df.head()
```

```python
x_data = df[[
 'Mean',
 'Variance',
 'Standard Deviation',
 'Entropy',
 'Skewness',
 'Kurtosis',
 'Contrast',
 'Energy',
 'ASM',
 'Homogeneity',
 'Dissimilarity',
 'Correlation',
 'Coarseness']]
y_data = df['Class']
```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set, 70% training and 30% test
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3,random_state=109) 
```

```python
#This is model generation

#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
model = GaussianNB()

#Train the model using the training sets
model.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = model.predict(X_test)

#This is to evaluate the model
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

**The classifier is correct 96% of the time.**
